```python
# 1

class BankAccount:
    def __init__(self, CustName, AccountNumber, balance, typeOfAccount, address):
        self.CustName = CustName
        self.AccountNumber = AccountNumber
        self.balance = balance
        self.typeOfAccount = typeOfAccount
        self.address = address
    def withdraw(self, cost):
        self.balance -= cost;
        print(self.balance)
    def deposit(self, cost):
        self.balance += cost;
        print(self.balance)
    def displayDetails(self):
        print(f"Account holder name is {self.CustName}")
        print(f"Account Number is {self.AccountNumber}")
        print(f"Balance is {self.balance}")
        print(f"typeOfAccount is {self.typeOfAccount}")
        print(f"address is {self.address}")
        
        
b1 = BankAccount("xyz", 123456, 12345, "saving", "msrit")
b2 = BankAccount("abc", 450835, 35874, "saving", "msrit2")

b1.withdraw(2345)
b1.displayDetails()

b2.deposit(232)
b2.displayDetails()
```

```python
# 3
class FourthSem:
    def __init__(self, roll, test1, test2, test3):
        self.RollNum = roll
        self.Test1Marks = test1
        self.Test2Marks = test2
        self.Test3Marks = test3

    def classAverage(self):
        first = sum(self.Test1Marks) / len(self.Test1Marks)
        print(f"Test 1 Marks Average of Class: {first}")

        second = sum(self.Test2Marks) / len(self.Test2Marks)
        print(f"Test 2 Marks Average of Class: {second}")

        third = sum(self.Test3Marks) / len(self.Test3Marks)
        print(f"Test 3 Marks Average of Class: {third}")

    def studentAverage(self, index):
        average = (self.Test1Marks[index] + self.Test2Marks[index] + self.Test3Marks[index]) / 3
        return average

    def firstLast(self):
        print("Top 5 scores for every test:")
        print(f"Test 1: {sorted(self.Test1Marks, reverse=True)[:5]}")
        print(f"Test 2: {sorted(self.Test2Marks, reverse=True)[:5]}")
        print(f"Test 3: {sorted(self.Test3Marks, reverse=True)[:5]}")

        print("Last 5 scores for every test:")
        print(f"Test 1: {sorted(self.Test1Marks)[:5]}")
        print(f"Test 2: {sorted(self.Test2Marks)[:5]}")
        print(f"Test 3: {sorted(self.Test3Marks)[:5]}")

rollno = [int(i) for i in input("Enter the roll nos(20 numbers): ").split()]
test1 = [int(i) for i in input("Enter test1 marks (for each student): ").split()]
test2 = [int(i) for i in input("Enter test2 marks (for each student): ").split()]
test3 = [int(i) for i in input("Enter test3 marks (for each student): ").split()]

cse = FourthSem(rollno, test1, test2, test3)

cse.classAverage()
cse.firstLast()
print("Average marks for student at index 5:", cse.studentAverage(3))
```

```python
# 4
class TimeClass:
    def __init__(self, hours=0, minutes=0, seconds=0):
        self.hours = int(hours)
        self.minutes = int(minutes)
        self.seconds = int(seconds)

    def displayTime(self):
        print(f"{self.hours:02d}:{self.minutes:02d}:{self.seconds:02d}")

    def addTime(self, other_time):
        total_seconds = self.countSeconds() + other_time.countSeconds()
        hours = total_seconds // 3600
        minutes = (total_seconds % 3600) // 60
        seconds = total_seconds % 60
        print(f"Sum of times: {hours:02d}:{minutes:02d}:{seconds:02d}")

    def subtractTime(self, other_time):
        total_seconds = abs(self.countSeconds() - other_time.countSeconds())
        hours = total_seconds // 3600
        minutes = (total_seconds % 3600) // 60
        seconds = total_seconds % 60
        print(f"Difference of times: {hours:02d}:{minutes:02d}:{seconds:02d}")

    def countSeconds(self):
        return self.hours * 3600 + self.minutes * 60 + self.seconds



time1 = TimeClass(2, 30, 45)
time2 = TimeClass(1, 15, 20)

print("Time 1:")
time1.displayTime()
print("Time 2:")
time2.displayTime()

time1.addTime(time2)
time1.subtractTime(time2)
```

```python
# 5
class Invoice:
    def __init__(self):
        self.items = []
        self.prices = []
    
    def add_item(self, item, price):
        self.items.append(item)
        self.prices.append(price)
    
    def calculate_tax(self, item_type):
        if item_type == 1:
            return 0.05
        elif item_type == 2:
            return 0.12
        elif item_type == 3:
            return 0.18
        elif item_type == 4:
            return 0.28
        else:
            return 0 
    
    def print_invoice(self):
        total = sum(self.prices)
        total_tax = sum(price * self.calculate_tax(item_type) for item_type, price in zip(self.items, self.prices))
        final_bill = total + total_tax
        
        print("Invoice:")
        print("Item  Price  GST")
        for i in range(len(self.items)):
            print(f"{self.items[i]}     ${self.prices[i]}    {self.calculate_tax(self.items[i]) * 100}%")
        print("----------------------")
        print(f"Total: ${total}")
        print(f"Total Tax: ${total_tax}")
        print("----------------------")
        print(f"Final Bill: ${final_bill}")



invoice = Invoice()
invoice.add_item(1, 100) 
invoice.add_item(2, 200)  
invoice.add_item(3, 150) 
invoice.add_item(4, 300)  

invoice.print_invoice()
```

```python
# 6
class Shape:
    def __init__(self, side1=0, side2=0):
        self.side1 = side1
        self.side2 = side2

    def calculateArea(self):
        pass  

    def displayDetails(self):
        pass  


class Rectangle(Shape):
    def __init__(self, length=0, width=0):
        super().__init__(length, width)

    def calculateArea(self):
        return self.side1 * self.side2

    def displayDetails(self):
        print("Rectangle:")
        print("Length:", self.side1)
        print("Width:", self.side2)
        print("Area:", self.calculateArea())


class Triangle(Shape):
    def __init__(self, base=0, height=0):
        super().__init__(base, height)

    def calculateArea(self):
        return 0.5 * self.side1 * self.side2

    def displayDetails(self):
        print("Triangle:")
        print("Base:", self.side1)
        print("Height:", self.side2)
        print("Area:", self.calculateArea())



rectangle = Rectangle(5, 4)
rectangle.displayDetails()

triangle = Triangle(3, 6)
triangle.displayDetails()
```

```python
# 7

class Quadratic:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    def find_roots(self):
        discriminant = self.b**2 - 4 * self.a * self.c
        if discriminant > 0:
            root1 = (-self.b + discriminant ** 0.5) / (2 * self.a)
            root2 = (-self.b - discriminant ** 0.5) / (2 * self.a)
            print("Two distinct real roots:")
            print("Root 1:", root1)
            print("Root 2:", root2)
        elif discriminant == 0:
            root = -self.b / (2 * self.a)
            print("One real root (repeated):")
            print("Root:", root)
        else:
            real_part = -self.b / (2 * self.a)
            imaginary_part = (-discriminant) ** 0.5 / (2 * self.a)
            print("Two complex roots:")
            print("Root 1:", real_part, "+", imaginary_part, "i")
            print("Root 2:", real_part, "-", imaginary_part, "i")
            
eq1 = Quadratic(1, -3, 2)
print("Equation: x^2 - 3x + 2 = 0")
eq1.find_roots()
eq2 = Quadratic(1, 2, 1)
print("\nEquation: x^2 + 2x + 1 = 0")
eq2.find_roots()
eq3 = Quadratic(1, -2, 5)
print("\nEquation: x^2 - 2x + 5 = 0")
eq3.find_roots()
    
```

```python
# 8

class Stack:
    arr =[]
    
    def isEmpty(self, arr):
        return (len(self.arr) == 0)

    def push(self, value):
        self.arr.append(value)
        return 

    def pop(self):
        if self.isEmpty(self.arr):
            print("Stack empty")
            return -1
        return self.arr.pop()

    def peek(self):
        if self.isEmpty(self.arr):
            print("Stack empty")
            return -1
        return self.arr[-1]

class Queue :
    arr = []
    
    def isEmpty(self):
        return (len(self.arr) == 0)
    
    def enqueue(self, value):
        self.arr.append(value)
    
    def dequeue(self):
        if self.isEmpty():
            return -1
        res = self.arr[0]
        self.arr = self.arr[1:]
        return res
    
    def peek(self):
        if self.isEmpty():
            return -1
        return self.arr[0]


s =  Stack()
q = Queue()
s.push(5)
q.enqueue(7)
s.push(6)
q.enqueue(8)
print(s.peek())
print(q.peek())
print(s.pop())
print(q.dequeue())
```
