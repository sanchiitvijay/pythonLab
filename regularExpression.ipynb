```python
# Develop a python program to count all the occurrences of vowels, consonants and
# digits from the given text using Regular expressions.

import re

def count_occurrences(text):
    vowels = re.findall(r'[aeiouAEIOU]', text)
    consonants = re.findall(r'[bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ]', text)
    digits = re.findall(r'\d', text)
    
    return len(vowels), len(consonants), len(digits)

text = input("Enter the text: ")
vowels_count, consonants_count, digits_count = count_occurrences(text)
print(f"Vowels: {vowels_count}, Consonants: {consonants_count}, Digits: {digits_count}")
```

```python
# Develop a python program to match all the occurrences of a word which has an "A/a"
# followed by
# i. Zero or more b's
# ii. Zero or one b
# iii. One and more b's
# iv. Four consecutive b's followed by a “c”


import re

def match_patterns(text):
    patterns = {
        "A/a followed by zero or more b's": r'\b[Aa]b*\b',
        "A/a followed by zero or one b": r'\b[Aa]b?\b',
        "A/a followed by one or more b's": r'\b[Aa]b+\b',
        "A/a followed by four b's and a c": r'\b[Aa]bbbbc\b'
    }
    
    matches = {desc: re.findall(pattern, text) for desc, pattern in patterns.items()}
    return matches

text = input("Enter the text: ")
matches = match_patterns(text)
for desc, match in matches.items():
    print(f"{desc}: {match}")
```

```python
# Develop a python program to read an USN and check if the USN starts with "1MS"
# and ends with "0". The length of the USN must be strictly 10. Display the appropriate
# results

import re

def validate_usn(usn):
    if re.fullmatch(r'1MS.{6}0', usn):
        return "Valid USN"
    else:
        return "Invalid USN"

usn = input("Enter the USN: ")
print(validate_usn(usn))
```

```python
# Develop a python program to read 20 random numbers. Display all the odd numbers
# from this list which are of length 2 and 4.

import random

def filter_odd_numbers(numbers):
    odd_numbers = [num for num in numbers if num % 2 != 0]
    result = [num for num in odd_numbers if len(str(num)) in [2, 4]]
    return result

random_numbers = []
for _ in range(20):
    num = int(input("Enter a random number: "))
    random_numbers.append(num)

print("Random numbers:", random_numbers)
filtered_numbers = filter_odd_numbers(random_numbers)
print("Filtered odd numbers with length 2 and 4:", filtered_numbers)
```

```python
# Develop a python program to search word in a string and display the location within
# the original string where the pattern occurs. If the pattern occurs multiple times, then
# display all the locations. Replace the pattern with another word.

import re

def search_and_replace(text, pattern, replacement):
    locations = [m.start() for m in re.finditer(pattern, text)]
    replaced_text = re.sub(pattern, replacement, text)
    return locations, replaced_text

text = input("Enter the text: ")
pattern = input("Enter the pattern to search: ")
replacement = input("Enter the replacement word: ")

locations, replaced_text = search_and_replace(text, pattern, replacement)
print("Locations:", locations)
print("Replaced text:", replaced_text)
```

```python
# Develop a python program to read a C-program and display all the keywords. Display
# the keywords and their locations. Search for the "int, float, while, for, if, if else, else,
# switch, printf and scanf" keywords in the program

import re

def find_keywords(c_program):
    keywords = ["int", "float", "while", "for", "if", "else", "switch", "printf", "scanf"]
    keyword_locations = {}
    
    for keyword in keywords:
        locations = [m.start() for m in re.finditer(r'\b' + keyword + r'\b', c_program)]
        if locations:
            keyword_locations[keyword] = locations
    
    return keyword_locations

print("Enter the C program (terminate with an empty line):")
lines = []
while True:
    line = input()
    if line == "":
        break
    lines.append(line)
c_program = "\n".join(lines)

keyword_locations = find_keywords(c_program)
print("Keyword locations:", keyword_locations)
```

```python
# Develop a python program to read convert a date of yyyy-mm-dd format to dd-mmyyyy format.


def convert_date_format(date_str):
    return '-'.join(date_str.split('-')[::-1])

date_str = input("Enter the date in yyyy-mm-dd format: ")
converted_date = convert_date_format(date_str)
print("Converted date:", converted_date)
```

```python
# Develop a python program to find all three, four and five characters long word in a
# string
import re

def find_short_words(text):
    return re.findall(r'\b\w{3,5}\b', text)

text = input("Enter the text: ")
short_words = find_short_words(text)
print("Three to five characters long words:", short_words)
    
```

```python
# Develop a python program to find all the urls in given string.

import re

def find_urls(text):
    url_pattern = r'https?://\S+|www\.\S+'
    return re.findall(url_pattern, text)

text = input("Enter the text containing URLs: ")
urls = find_urls(text)
print("URLs found:", urls)
```
